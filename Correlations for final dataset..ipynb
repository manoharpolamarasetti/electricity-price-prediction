```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\t10mp22\OneDrive - University of Aberdeen\Project\final dataset(historic_demand_prices).csv")
df['DATETIME'] = pd.to_datetime(df['DATETIME'], dayfirst=True)  # Convert DATETIME column to datetime format

# Filter for years 2009 to 2015
filtered_df = df[(df['DATETIME'].dt.year >= 2009) & (df['DATETIME'].dt.year <= 2023)]

# Expanded list of columns to be excluded from the heatmap
columns_to_exclude = [
    'GENERATION_perc', 'GAS_perc', 'COAL_perc', 'NUCLEAR_perc', 'WIND_perc', 
    'HYDRO_perc', 'IMPORTS_perc', 'BIOMASS_perc', 'OTHER_perc', 'SOLAR_perc', 'STORAGE_perc', 
    'LOW_CARBON_perc', 'ZERO_CARBON_perc', 'RENEWABLE_perc', 'FOSSIL_perc'
]

# Exclude specified columns and drop any other non-numeric columns that cannot be correlated
heatmap_df = filtered_df.drop(columns=columns_to_exclude).select_dtypes(include=['number'])

# Calculate the correlation matrix for the remaining columns
correlation_matrix = heatmap_df.corr()

# Generate a heatmap based on the correlation matrix, including correlation values
plt.figure(figsize=(20, 18))  # Adjust the size as needed
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".02f")
plt.title('Correlation Heatmap with Values (2009-2023)')

plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
plt.show()

correlation_matrix
```
